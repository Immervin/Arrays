```python
n=int(input())
lst=list(map(int,input().split()))
counter=0
count=0
result=[]

while counter<n:
  if lst[count]>lst[counter]:
    result.append(lst[counter])
    count+=1
    # print('count '+str(count))

  else:
    counter+=1
    if counter==n:
      for i in range(n-count):
        if count<counter:
          result.append(-1)
        else:
          break

for i,j in enumerate(result):
  if i==n-1:
    print(j,end='')
  else:
    print(j,end=" ")




```

```python
n,m=map(int,input().split())
val=list(map(int,input().split()))
loop=len(val)
result=[]

for i in range(loop):
  diff=abs(val[i]-val[(i+1)%loop])
  if diff>m:
    result.append(1)
  else:
    result.append(0)

print(*result)

```
